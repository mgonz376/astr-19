```python
import numpy as np
import matplotlib.pyplot as plt

#generate 1000 random numbers uniformly distributed between [0,1]
random_numbers = np.random.uniform(0, 1, 1000)

#histogram with 100 bins
plt.figure(figsize=(10,8))
plt.hist(random_numbers, bins=100, color='blue', alpha=0.7)

#axes labels
plt.title('Histogram of 1000 Uniformly Distributed Random Numbers')
plt.xlabel('values')
plt.ylabel('frequency')

plt.savefig('session_9histo.pdf')
plt.show
```
