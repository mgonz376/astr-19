```python
import numpy as np 
import matplotlib.pyplot as plt

#array ranging from [0,1] in 100 steps
x = np.linspace(0, 1, 100)

#functions that return sin and cos 
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
#creating a multipanel plot 

plt.figure(figsize= (14,8))

#sin panel
plt.subplot(1,2,1)
plt.plot(x, sin_function(x), label='sin(x)', color='blue')
plt.title('sin(x) vs x')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
plt.legend()

#cos panel
plt.subplot(1,2,2)
plt.plot(x, cos_function(x), label= 'cos(x)', color='red')
plt.title('cos(x) vs x')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
plt.legend()

plt.savefig('session8.pdf')

plt.show()
```
