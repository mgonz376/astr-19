```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create an array x ranging form [0,1] in 100 steps 
x=np.linspace(0,1,100)
```

```python
#define function that returns exp(x)

def exp_function(x):
    return np.exp(x)

y = exp_function(x)

#plotting the graph
plt.figure(figsize=(5,5))
plt.plot(x, y, label= 'Session 7 Jupyter Notebook', color='red')
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.legend()
plt.grid()

plt.savefig('session_7.pdf') #save figure

plt.show()
```
